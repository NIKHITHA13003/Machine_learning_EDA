### Here the dataframe split in to train and test data with testsize is 0.2 that is 20% 
* my total data was 400 , train data contains 320 rows that is 80% and test data contain 20% that is 80 rows

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x_train,x_test)
```

```python
x_train_scaler=scaler.transform(x_train)
x_train_scaler
```

```python
x_test_scaler=scaler.transform(x_test)
x_test_scaler
```

```python
x_train_scaler=pd.DataFrame(x_train_scaler,columns=x_train.columns)
x_train_scaler
```

```python
x_test_scaler=pd.DataFrame(x_test_scaler,columns=x_test.columns)
x_test_scaler
```

```python
np.round(x_train_scaler.describe())
```

```python
np.round(x_test_scaler.describe())
```

### observe this few things here the standard deviation and mean value will change because we scale up

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title("Before scaling")
ax1.scatter(x_train["Age"],x_train["EstimatedSalary"])
ax2.set_title("After scaling")
ax2.scatter(x_train_scaler["Age"],x_train_scaler["EstimatedSalary"],color="green")
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title("Before scaling")
sns.kdeplot(df["Age"],ax=ax1,color="green")
sns.kdeplot(df["EstimatedSalary"],ax=ax1,color="orange")
ax2.set_title("After scaling")
sns.kdeplot(x_train_scaler["Age"],ax=ax2,color="green")
sns.kdeplot(x_train_scaler["EstimatedSalary"],ax=ax2,color="orange")
plt.show()
```
